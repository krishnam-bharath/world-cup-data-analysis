```python
import pandas as pd
from google.colab import files

# Upload the CSV file
uploaded = files.upload()

# Replace with your actual file name if different
df = pd.read_csv("WorldCupMatches.csv")

# Data Cleaning
df = df.dropna(subset=["Year"])
df["Year"] = df["Year"].astype(int)

df["Attendance"] = pd.to_numeric(df["Attendance"], errors="coerce")
df["Home Team Goals"] = pd.to_numeric(df["Home Team Goals"], errors="coerce")
df["Away Team Goals"] = pd.to_numeric(df["Away Team Goals"], errors="coerce")
```

```python
matches_per_year = df.groupby("Year").size().reset_index(name="Total Matches")
print(matches_per_year)
```

```python
attendance_per_year = df.groupby("Year")["Attendance"].sum().reset_index(name="Total Attendance")
print(attendance_per_year)
```

```python
avg_home_goals = df.groupby("Year")["Home Team Goals"].mean().reset_index(name="Average Home Goals")
print(avg_home_goals)
```

```python
avg_away_goals = df.groupby("Year")["Away Team Goals"].mean().reset_index(name="Average Away Goals")
print(avg_away_goals)
```

```python
matches_per_stage = df.groupby("Stage").size().reset_index(name="Matches")
print(matches_per_stage)
```

```python
matches_per_city = df.groupby("City").size().reset_index(name="Matches")
print(matches_per_city)
```

```python
matches_per_stadium = df.groupby("Stadium").size().reset_index(name="Matches")
print(matches_per_stadium)
```

```python
matches_per_referee = (
    df.groupby("Referee")
      .size()
      .reset_index(name="Matches Officiated")
      .sort_values(by="Matches Officiated", ascending=False)
)

print(matches_per_referee)
```

```python
matches_per_home_team = (
    df.groupby("Home Team Name")
      .size()
      .reset_index(name="Home Matches")
      .sort_values(by="Home Matches", ascending=False)
)

print(matches_per_home_team)
```

```python
matches_per_away_team = (
    df.groupby("Away Team Name")
      .size()
      .reset_index(name="Away Matches")
      .sort_values(by="Away Matches", ascending=False)
)

print(matches_per_away_team)
```
